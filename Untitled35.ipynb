```python
import random
import string
import secrets
from datetime import datetime

def generate_public_code(base: str, discount: int) -> str:
    # Example: "WELCOME10"
    return f"{base.upper()}{discount}"

def generate_seasonal_code(event_name: str, year: int | None = None) -> str:
    year = year or datetime.utcnow().year
    return f"{event_name.upper()}{year}"

def generate_limited_time_code(base: str, discount: int, length: int = 3) -> str:
    suffix = ''.join(random.choices(string.ascii_uppercase + string.digits, k=length))
    return f"{base.upper()}{discount}{suffix}"

def generate_personalized_code(referrer: str, discount: int) -> str:
    return f"{referrer.upper()}{discount}"

def generate_secure_code(prefix: str = "COUPON", length: int = 6) -> str:
    token = secrets.token_hex(length // 2).upper()
    return f"{prefix}-{token}"

def generate_coupon_code(
    code_type: str,
    base: str = "",
    discount: int = 0,
    referrer: str = ""
) -> str:
    if code_type == "public":
        return generate_public_code(base, discount)
    elif code_type == "seasonal":
        return generate_seasonal_code(base)
    elif code_type == "limited":
        return generate_limited_time_code(base, discount)
    elif code_type == "personalized":
        return generate_personalized_code(referrer, discount)
    elif code_type == "random":
        return generate_secure_code(base)
    else:
        raise ValueError("Unknown coupon type")



```

```python
from PIL import Image, ImageDraw, ImageFilter
import requests
import io
import matplotlib.pyplot as plt

# Constants
DPI = 300
MM_TO_INCH = 25.4
CIRCLE_MM = 20
CIRCLE_PX = round(DPI * CIRCLE_MM / MM_TO_INCH)
ROWS = 3
COLS = 4
MARGIN_MM = 15
SPACING_MM = 25

# Convert mm to pixels
margin_px = round(MARGIN_MM * DPI / MM_TO_INCH)
spacing_px = round(SPACING_MM * DPI / MM_TO_INCH)
canvas_width = margin_px * 2 + (COLS - 1) * spacing_px + COLS * CIRCLE_PX
canvas_height = margin_px * 2 + (ROWS - 1) * spacing_px + ROWS * CIRCLE_PX

# Image loader
def load_image(url_or_path: str) -> Image.Image:
    if url_or_path.startswith("http"):
        response = requests.get(url_or_path)
        img = Image.open(io.BytesIO(response.content))
    else:
        img = Image.open(url_or_path)
    return img.convert("RGBA")

# Generate concentric image
def create_concentric_image(img1: Image.Image, img2: Image.Image) -> Image.Image:
    inner_px = CIRCLE_PX // 2
    img1_resized = img1.resize((inner_px, inner_px), Image.Resampling.LANCZOS)
    img2_resized = img2.resize((CIRCLE_PX, CIRCLE_PX), Image.Resampling.LANCZOS)

    # Masks
    mask_inner = Image.new("L", (inner_px, inner_px), 0)
    draw_inner = ImageDraw.Draw(mask_inner)
    draw_inner.ellipse((0, 0, inner_px, inner_px), fill=255)

    mask_outer = Image.new("L", (CIRCLE_PX, CIRCLE_PX), 0)
    draw_outer = ImageDraw.Draw(mask_outer)
    draw_outer.ellipse((0, 0, CIRCLE_PX, CIRCLE_PX), fill=255)
    draw_outer.ellipse(
        (CIRCLE_PX//2 - inner_px//2, CIRCLE_PX//2 - inner_px//2,
         CIRCLE_PX//2 + inner_px//2, CIRCLE_PX//2 + inner_px//2),
        fill=0,
    )
    mask_outer = mask_outer.filter(ImageFilter.GaussianBlur(0.8))

    # Compositing
    base = Image.new("RGBA", (CIRCLE_PX, CIRCLE_PX), (0, 0, 0, 0))
    inner = Image.new("RGBA", (CIRCLE_PX, CIRCLE_PX), (0, 0, 0, 0))

    base.paste(img2_resized, (0, 0), mask_outer)
    offset = CIRCLE_PX // 2 - inner_px // 2
    inner.paste(img1_resized, (offset, offset), mask_inner)

    return Image.alpha_composite(base, inner)

# Load images
image1 = load_image("https://t3.ftcdn.net/jpg/06/05/37/40/240_F_605374009_hEUHatmKPzuHTIacg7rLneAgnLHUgegM.jpg")
image2 = load_image("https://t4.ftcdn.net/jpg/03/03/62/45/240_F_303624505_u0bFT1Rnoj8CMUSs8wMCwoKlnWlh5Jiq.jpg")

# Create a single concentric circle image
circle_image = create_concentric_image(image1, image2)

# Create the full canvas
canvas = Image.new("RGBA", (canvas_width, canvas_height), (255, 255, 255, 255))

# Paste 12 images in a 4x3 grid
for row in range(ROWS):
    for col in range(COLS):
        x = margin_px + col * (CIRCLE_PX + spacing_px)
        y = margin_px + row * (CIRCLE_PX + spacing_px)
        canvas.paste(circle_image, (x, y), circle_image)

# Show result
plt.figure(figsize=(10, 8))
plt.imshow(canvas)
plt.axis("off")
plt.title("43 Grid of Concentric Circle Images (20mm)")
plt.show()

# Save if needed
# canvas.save("grid_concentric.png", dpi=(300, 300))
```

```python
from PIL import Image, ImageDraw, ImageFilter, ImageOps, ImageEnhance
import requests
import io
import matplotlib.pyplot as plt
import numpy as np

# Constants (redefined for clarity within this cell)
DPI = 300
MM_TO_INCH = 25.4
INNER_MM = 10
OUTER_MM = 20
SHADOW_OFFSET = 10  # Pixels
SHADOW_BLUR_RADIUS = 10
BORDER_WIDTH = 5  # Pixels
BORDER_COLOR = "black"

# Convert mm to pixels
inner_px = round(DPI * INNER_MM / MM_TO_INCH)
outer_px = round(DPI * OUTER_MM / MM_TO_INCH)

# Function to load image (redefined for clarity within this cell)
def load_image(path_or_url: str) -> Image.Image:
    if path_or_url.startswith("http"):
        response = requests.get(path_or_url)
        img = Image.open(io.BytesIO(response.content))
    else:
        img = Image.open(path_or_url)
    return img.convert("RGBA")

# Function to preprocess images based on their position in the concentric design
def preprocess_image(img: Image.Image, position: str) -> Image.Image:
    """
    Applies preprocessing steps to an image based on its intended position
    (inner or outer) in a concentric design.

    Args:
        img: The input Pillow Image object.
        position: A string, either "inner" or "outer".

    Returns:
        The preprocessed Pillow Image object.
    """
    enhanced_img = img.copy()

    if position == "inner":
        # Enhance contrast and saturation for the inner image to make it stand out
        enhancer_contrast = ImageEnhance.Contrast(enhanced_img)
        enhanced_img = enhancer_contrast.enhance(1.5) # Increase contrast by 50%

        enhancer_color = ImageEnhance.Color(enhanced_img)
        enhanced_img = enhancer_color.enhance(1.3) # Increase saturation by 30%

        print("Preprocessing: Applied contrast and saturation enhancement to inner image.")

    elif position == "outer":
        # Slightly reduce saturation and add a subtle blur for the outer ring
        enhancer_color = ImageEnhance.Color(enhanced_img)
        enhanced_img = enhancer_color.enhance(0.8) # Reduce saturation by 20%

        enhanced_img = enhanced_img.filter(ImageFilter.GaussianBlur(radius=2)) # Add subtle blur

        print("Preprocessing: Applied saturation reduction and subtle blur to outer image.")

    return enhanced_img


# Create concentric image with improved blending using refined gradient mask and effects
def create_concentric_image_with_effects(img1: Image.Image, img2: Image.Image, add_shadow: bool = True, add_border: bool = True) -> Image.Image:
    # Resize both images
    img1_resized = img1.resize((inner_px, inner_px), Image.Resampling.LANCZOS)
    img2_resized = img2.resize((outer_px, outer_px), Image.Resampling.LANCZOS)

    # Create inner mask
    mask_inner = Image.new("L", (inner_px, inner_px), 0)
    draw_inner = ImageDraw.Draw(mask_inner)
    draw_inner.ellipse((0, 0, inner_px, inner_px), fill=255)

    # Create outer mask with a radial gradient
    mask_outer = Image.new("L", (outer_px, outer_px), 0)
    y, x = np.indices((outer_px, outer_px))
    center_x, center_y = outer_px // 2, outer_px // 2
    distance_from_center = np.sqrt((x - center_x)**2 + (y - center_y)**2)

    inner_radius = inner_px / 2
    outer_radius = outer_px / 2

    gradient_start = inner_radius
    gradient_end = outer_radius

    alpha_values = np.interp(distance_from_center, [gradient_start, gradient_end], [0, 255])
    alpha_values = np.clip(alpha_values, 0, 255).astype(np.uint8)

    mask_outer = Image.fromarray(alpha_values, 'L')
    mask_outer = mask_outer.filter(ImageFilter.GaussianBlur(radius=1))

    # Canvases
    outer_canvas = Image.new("RGBA", (outer_px, outer_px), (0, 0, 0, 0))
    inner_canvas = Image.new("RGBA", (outer_px, outer_px), (0, 0, 0, 0))

    outer_canvas.paste(img2_resized, (0, 0), mask_outer)
    offset = outer_px // 2 - inner_px // 2
    inner_canvas.paste(img1_resized, (offset, offset), mask_inner)

    result = Image.alpha_composite(outer_canvas, inner_canvas)

    # Add drop shadow
    if add_shadow:
        shadow_canvas_size = (outer_px + SHADOW_OFFSET * 2 + SHADOW_BLUR_RADIUS * 2,
                              outer_px + SHADOW_OFFSET * 2 + SHADOW_BLUR_RADIUS * 2)
        shadow_canvas = Image.new("RGBA", shadow_canvas_size, (0, 0, 0, 0))

        # Create shadow mask from the composite image's alpha channel
        shadow_mask = result.getchannel("A").copy()

        # Calculate the position to paste the original result image onto the larger shadow canvas
        paste_x_orig = SHADOW_OFFSET + SHADOW_BLUR_RADIUS
        paste_y_orig = SHADOW_OFFSET + SHADOW_BLUR_RADIUS

        # Create a black fill with the shadow mask
        shadow = Image.new("RGBA", result.size, (0, 0, 0, 128)) # Semi-transparent black
        shadow.putalpha(shadow_mask)
        shadow = shadow.filter(ImageFilter.GaussianBlur(radius=SHADOW_BLUR_RADIUS))

        # Paste shadow onto the canvas at the shadow offset
        shadow_canvas.paste(shadow, (SHADOW_OFFSET, SHADOW_OFFSET), shadow)

        # Paste the original image on top
        shadow_canvas.paste(result, (paste_x_orig, paste_y_orig), result)
        result = shadow_canvas

    # Add border
    if add_border:
        current_width, current_height = result.size
        border_canvas_size = (current_width + BORDER_WIDTH * 2, current_height + BORDER_WIDTH * 2)
        border_canvas = Image.new("RGBA", border_canvas_size, (0, 0, 0, 0))
        draw_border = ImageDraw.Draw(border_canvas)

        # Draw the border rectangle
        draw_border.rectangle([(0, 0), (border_canvas_size[0] - 1, border_canvas_size[1] - 1)], outline=BORDER_COLOR, width=BORDER_WIDTH)

        # Paste the image onto the border canvas
        border_canvas.paste(result, (BORDER_WIDTH, BORDER_WIDTH), result)
        result = border_canvas

    return result

# Load images
image1_raw = load_image("https://t3.ftcdn.net/jpg/06/05/37/40/240_F_605374009_hEUHatmKPzuHTIacg7rLneAgnLHUgegM.jpg")
image2_raw = load_image("https://t4.ftcdn.net/jpg/03/03/62/45/240_F_303624505_u0bFT1Rnoj8CMUSs8wMCwoKlnWlh5Jiq.jpg")

# Preprocess images
image1_preprocessed = preprocess_image(image1_raw, "inner")
image2_preprocessed = preprocess_image(image2_raw, "outer")

# Generate and display the concentric image with effects using preprocessed images
final_image_preprocessed = create_concentric_image_with_effects(image1_preprocessed, image2_preprocessed, add_shadow=True, add_border=True)

plt.figure(figsize=(6, 6))
plt.imshow(final_image_preprocessed)
plt.axis("off")
plt.title("Concentric Image with Preprocessing, Drop Shadow, and Border")
plt.show()
```
