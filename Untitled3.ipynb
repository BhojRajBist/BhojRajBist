```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read the shapefile
shapefile_path = "/content/hermes_NPL_new_wgs_1 (2).shp"
gdf = gpd.read_file(shapefile_path)

# Generate random colors based on the number of unique provinces
unique_provinces = gdf['PROVINCE'].unique()
num_unique_provinces = len(unique_provinces)
random_colors = np.random.rand(num_unique_provinces, 3)

# Create a dictionary to map province IDs to random colors
province_color_mapping = {tuple([province_id]): tuple(random_colors[i]) for i, province_id in enumerate(unique_provinces)}

# Add a new column 'color' to the GeoDataFrame based on province ID
gdf['color'] = gdf['PROVINCE'].apply(lambda x: province_color_mapping.get((x,), None))

# Display the GeoDataFrame's geometry and attributes
print("Geometry:")
print(gdf.geometry.head())

print("\nAttributes:")
print(gdf.head())

# Plot the GeoDataFrame with random colors
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='color', ax=ax, legend=True)

# Customize the legend
handles = [plt.Line2D([0], [0], marker='o', color='w', label=str(province_id), markersize=10, markerfacecolor=color) for province_id, color in province_color_mapping.items()]
ax.legend(handles=handles, title='Province Legend')

# Show the map
plt.show()


```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import to_hex

# Step 1: Read the shapefile
shapefile_path = "/content/hermes_NPL_new_wgs_1 (2).shp"
gdf = gpd.read_file(shapefile_path)
unique_provinces = gdf['PROVINCE'].unique()
num_unique_provinces = len(unique_provinces)
random_colors = np.random.rand(num_unique_provinces, 3)

# Create a dictionary to map province IDs to random colors and PR_NAME
province_color_mapping = {province_id: {'color': random_colors[i], 'name': name} for i, (province_id, name) in enumerate(zip(unique_provinces, gdf['PR_NAME']))}

# Define a function to get hex color from RGB values
def rgb_to_hex(rgb):
    return "#{:02x}{:02x}{:02x}".format(int(rgb[0]*255), int(rgb[1]*255), int(rgb[2]*255))

# Add a new column 'color' to the GeoDataFrame based on PROVINCE
gdf['color'] = gdf['PROVINCE'].apply(lambda x: province_color_mapping.get(x, {}).get('color', None))

# Convert the color column to hex format
gdf['color_hex'] = gdf['color'].apply(lambda x: rgb_to_hex(x) if x is not None else None)

# Display the GeoDataFrame's geometry and attributes
print("Geometry:")
print(gdf.geometry.head())

print("\nAttributes:")
print(gdf.head())

# Plot the GeoDataFrame with random colors
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='color_hex', ax=ax, legend=True)

# Customize the legend to display PR_NAME
handles = [plt.Line2D([0], [0], marker='o', color='w', label=data['name'], markersize=10, markerfacecolor=to_hex(data['color'])) for province_id, data in province_color_mapping.items()]
ax.legend(handles=handles, title='Province Legend')

# Show the map
plt.show()


```
