```python
import geopandas as gpd

# Load the Parcel Shapefile
parcels = gpd.read_file("/content/ShapeFiles/parcel_Barkalpur_9ga.shp")

# Load the Land Use Shapefile
land_use = gpd.read_file("/content/ShapeFiles/Agriculture.shp")

# Ensure both datasets have the same CRS
if parcels.crs != land_use.crs:
    land_use = land_use.to_crs(parcels.crs)

# Perform spatial intersection to split parcels based on land use
split_parcels = gpd.overlay(parcels, land_use, how='intersection')

# Save the output to a new shapefile
split_parcels.to_file("split_classified_parcels.shp")

print("Parcel splitting and classification completed successfully.")
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd

# Load Parcel and Land Use Shapefiles
parcels = gpd.read_file("/content/ShapeFiles/parcel_Barkalpur_9ga.shp")
land_use = gpd.read_file("/content/ShapeFiles/Agriculture.shp")
split_parcels = gpd.read_file("/content/split_classified_parcels.shp")  # Output from previous step

# Create a plot
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Plot Original Parcels
parcels.plot(ax=ax[0], edgecolor="black", facecolor="none")
ax[0].set_title("Original Parcels")

# Plot Land Use Zones
land_use.plot(ax=ax[1], edgecolor="red", facecolor="none", linestyle="--")
ax[1].set_title("Land Use Zones")

# Plot Split and Classified Parcels
split_parcels.plot(ax=ax[2], cmap="Set1", alpha=0.6, legend=True)
ax[2].set_title("Split and Classified Parcels")

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the original parcel shapefile
parcels = gpd.read_file("/content/ShapeFiles/parcel_Barkalpur_9ga.shp")

# Load the split and classified parcels
split_parcels = gpd.read_file("/content/split_classified_parcels.shp")

# Ensure both have the same CRS
if parcels.crs != split_parcels.crs:
    split_parcels = split_parcels.to_crs(parcels.crs)

# Create a plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot original parcels in black outlines
parcels.plot(ax=ax, edgecolor="black", facecolor="none", linewidth=1, label="Original Parcels")

# Plot split and classified parcels with color
split_parcels.plot(ax=ax, cmap="Set1", alpha=0.6, legend=True)

# Add title and legend
plt.title("Overlapping Original and Split Classified Parcels")
plt.legend()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the original parcels
parcels = gpd.read_file("/content/ShapeFiles/parcel_Barkalpur_9ga.shp")

# Load the split and classified parcels
split_parcels = gpd.read_file("/content/split_classified_parcels.shp")

# Ensure both datasets have the same CRS
if parcels.crs != split_parcels.crs:
    split_parcels = split_parcels.to_crs(parcels.crs)

# Create a figure
fig, ax = plt.subplots(figsize=(10, 10))

# Plot original parcels in light gray (background)
parcels.plot(ax=ax, edgecolor="gray", facecolor="none", linewidth=0.8, label="Original Parcels")

# Plot split parcels with bold red outline and filled colors
split_parcels.plot(ax=ax, cmap="Set1", edgecolor="red", linewidth=1.5, alpha=0.7, legend=True)

# Add title and legend
plt.title("Highlighted Split Classified Parcels Over Original Parcels", fontsize=14)
plt.legend()
plt.show()
```
