```python
import requests
import os
import xarray as xr
import matplotlib.pyplot as plt

# Function to check if a file exists
def check_file_exists(url):
    response = requests.head(url)
    return response.status_code == 200

# Function to download a file from a URL
def download_file(url, output_path):
    response = requests.get(url, stream=True)
    if response.status_code == 200:
        with open(output_path, "wb") as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f"Downloaded: {output_path}")
        return True
    else:
        print(f"Failed to download {url} (Status Code: {response.status_code})")
        return False
```

```python
import requests

# Base URL
aws_base_url = "https://noaa-himawari8.s3.amazonaws.com/"

# Function to construct and check file URLs
def check_files(base_url, date, start_hour=0, end_hour=23):
    region = "AHI-L1b-FLDK"
    year, month, day = date[:4], date[4:6], date[6:8]
    channel = "B01"  # Change as needed
    resolution = "R10"  # Change as needed
    found_files = []

    for hour in range(start_hour, end_hour + 1):
        for minute in range(0, 60, 10):  # Himawari-8 provides data every 10 minutes
            hhmm = f"{hour:02d}{minute:02d}"
            file_path = f"{region}/{year}/{month}/{day}/{hhmm}/HS_H08_{date}{hhmm}_{channel}_FLDK_{resolution}_S0101.DAT.bz2"
            file_url = base_url + file_path

            response = requests.head(file_url)
            if response.status_code == 200:
                found_files.append(file_url)
                print(f"File found: {file_url}")
            else:
                print(f"File not found: {file_url}")

    return found_files

# Example date to check
date = "20240110"  # Replace with the desired date
found_files = check_files(aws_base_url, date)

print(f"Available files for {date}: {found_files}")
```

```python
import requests
import os

# Define the API endpoint and parameters
url = "https://www.jma.go.jp/jma/jma-eng/satellite/img/Himawari8"

# Define the parameters: specify the time, region, and channels.
params = {
    'date': '2019-07-31T07:00',  # Example date-time
    'latitude': 28.3949,  # Latitude of Nepal
    'longitude': 84.1240,  # Longitude of Nepal
    'channel': 'IR1',  # Example: Infrared channel
    'format': 'DAT'  # Data format (DAT or other available formats)
}

# Make a request to fetch the data
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Define the file path and save the data
    file_path = os.path.join('data', 'Himawari_20190731_0700_IR1.DAT')
    with open(file_path, 'wb') as f:
        f.write(response.content)
    print(f"Data downloaded successfully: {file_path}")
else:
    print("Failed to fetch data")
```
