```python
import matplotlib.pyplot as plt


# Data

categories = ['Total Students Appeared', 'Passed', 'Passed user of Examneer']

students = [96, 65, 29]


# Create a bar chart with all bars green

plt.figure(figsize=(8, 6))  # Set the figure size

plt.bar(categories, students, color='green')

plt.xlabel('')

plt.ylabel('Number of Students')

title = 'Student Performance in NEC Liscence Kartik 2080'

plt.title(title)


# Add data labels on top of the bars

for i, v in enumerate(students):

    plt.text(i, v + 1, str(v), ha='center', va='bottom', fontsize=12)  # Adjust position and font size


# Rotate x-axis labels for better readability

# plt.xticks(rotation=45, ha='right')


# Display the chart

plt.tight_layout()

plt.show()
```
