```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Define the original triangle coordinates
triangle = np.array([[0, 0], [1, 0], [0.5, 0.866]])

# Define rotation angle in radians
angle = np.pi / 4  # 45 degrees

# Define translation vector
translation = np.array([2, 2])

# Define scaling factors for x and y directions
scaling_factors = np.array([2, 0.5])

# Rotation matrix
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                            [np.sin(angle), np.cos(angle)]])

# Rotate the triangle coordinates
rotated_triangle = np.dot(triangle, rotation_matrix.T)

# Translate the rotated triangle
translated_triangle = rotated_triangle + translation

# Scale the translated triangle
scaled_triangle = translated_triangle * scaling_factors

# Plotting
fig, ax = plt.subplots()

original_poly = Polygon(triangle, fill=None, edgecolor='blue', label='Original')
rotated_poly = Polygon(rotated_triangle, fill=None, edgecolor='green', label='Rotated')
translated_poly = Polygon(translated_triangle, fill=None, edgecolor='red', label='Translated')
scaled_poly = Polygon(scaled_triangle, fill=None, edgecolor='purple', label='Scaled')

ax.add_patch(original_poly)
ax.add_patch(rotated_poly)
ax.add_patch(translated_poly)
ax.add_patch(scaled_poly)

ax.set_xlim(-2, 6)
ax.set_ylim(-2, 6)
ax.set_aspect('equal', adjustable='box')
ax.legend()

plt.show()
```
